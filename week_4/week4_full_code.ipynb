```python
#--- imports ---
import random
import time

#--- functions ---

def get_player_input(input_type, input_statement, max_allotted_time): 
    if input_type == 'timed': 
        start_time = time.time() 
        player_action = input(input_statement).upper() 
        end_time = time.time() 
        
        elapsed_time = end_time - start_time 
        if elapsed_time > max_allotted_time: 
            player_action = 'C' 
            
        return player_action

def rand_consequence():
       rand_num = random.randint(1, 100)
       if rand_num >= 50:
              return False
       else:
              return True

def chose_npc_action(): 

    action = random.randint(1, 3) 
    
    match action: 
          case 1: return 'H' 
          case 2: return 'L' 
          case 3: return 'B' 
          case _: 
            print(action) 
            return 0
          
def attack_results(attacker_action, defender_action, bad_status):
        
    action_code = attacker_action + defender_action
    attacker_damage_delt = 0

    if bad_status == True:
            attacker_damage_delt = random.randint(1, 20)
            return attacker_damage_delt
    else:
        match action_code: 
            case 'HH': print('high, high'); attacker_damage_delt = random.randint(0, 3) 
            case 'LL': print('low, low'); attacker_damage_delt = random.randint(0, 3) 
            case 'HL': print('high, low'); attacker_damage_delt = random.randint(1, 5)
            case 'LH': print('low, high'); attacker_damage_delt = random.randint(1, 5)
            case 'HB': print('high, block'); attacker_damage_delt = random.randint(1, 5) 
            case 'LB': print('low, block'); attacker_damage_delt = random.randint(1, 5) 
            case 'BH': print('Block, high'); attacker_damage_delt = 0 
            case 'BL': print('block, low'); attacker_damage_delt = 0  
            case 'BB': print('block, block'); attacker_damage_delt = 0 
            case _: print(f'missed something: {attacker_action}, {defender_action}'); attacker_damage_delt = 0 
        
        return attacker_damage_delt

def bridge_crossing():
    idk = True

    while idk:

        cross = input('Do you want to cross the Bridge of Death? (yes/no)').lower()

        if cross == 'yes':
            time.sleep(5)
            print('A troll appears and asks you three questions!')
            time.sleep(5)
            print(f'OF COURSE! you were able to answer the questions correctly!')
            time.sleep(5)
            print('How could you not?\nWhen they were questions such as "What is your favorite color"?')
            time.sleep(5)
            print('Although come to think of it, he did ask for you home address.....\nGuess you should expect some junk mail in the future...')
            idk = False
        elif cross == 'no':
            print('You wisely turn back')
            idk = False
        else:
            print('Please answer yes or no.')
        

def battle_function(choice_statement, consequences):

    player_health = 36
    npc_health = 36
    bad_status = False
    player_move = ''
    npc_action = ''
    whos_turn = ''

    print(f'Player Health: {player_health}\nNPC Health: {npc_health}')

    if consequences == True:
            whos_turn = 'npc'
            bad_status = True
    else:
            whos_turn = 'player'

    while player_health > 30 and npc_health > 30:

            if bad_status == False:
                player_move = input('Please take action! High swing: H, Low swing: L, Block: B').upper()
                npc_action = chose_npc_action() 
            
            if whos_turn == 'npc':
                    print(f'npc turn - player Health: {player_health}')
                    damage = attack_results(npc_action, player_move, bad_status)
                    player_health = player_health - damage
                    whos_turn = 'player'
                    print(f'Damage to player: {damage}')
            elif whos_turn == 'player':
                    print(f'player turn - npc Health: {npc_health}')
                    damage = attack_results(player_move, npc_action, bad_status)
                    npc_health = npc_health - damage
                    whos_turn = 'npc'
                    print(f'Damage to NPC: {damage}')
            
            bad_status = False

    if player_health <= 30:
            return 'npc wins'
    elif npc_health <= 30:
            return 'player wins'

#--- main ---
def main():
    
    print('Welcome to the Monty Python Adventure!\nYou are about to embark on a quest for the Holy Grail.\nBeware of dangerous encounters!')
    time.sleep(3)
    player_name = input("What is your name, brave adventurer? ")
    time.sleep(3)
    print(f'Welcome, Sir {player_name} of Pythonia!')
    time.sleep(3)
    print('You begin your journey through a dense forest abundant with wildlife.')
    time.sleep(3)
    print('You travel down a long forgotten path through the forest when suddenly...')
    time.sleep(5)
    print('A knight steps in front of you and exclaims! "None shall pass!"')

    player_action = get_player_input('timed', 'Will you? A. Fight the knight. or B. Reason with the Knight. Please type A or B. ', 10)

    results = 'start'
    match player_action: 
        case 'A': print("The knight draws his sword. It's time for battle!"); results = battle_function("The knight draws his sword. It's time for battle!", False)  
        case 'B': print("The knight scoffs at your attempt to reason and attacks anyway."); results = battle_function("The knight scoffs at your attempt to reason and attacks anyway.", True) 
        case _: print("You hesitate... and the knight charges!"); results = battle_function("You hesitate... and the knight charges!", rand_consequence())

    if results == 'npc wins':
        print(f'Sir {player_name} hobbles off, defeated in battle!')
        time.sleep(5)
        print(f'Sir {player_name} returns home, head hung low in defeat.')
    elif results == 'player wins':
        print('The player exclaims:\nI have won the battle!\nThe Knight replies, I am not defeated Tis but a flesh wound!')
        time.sleep(3)
        print('The player replies, You have no Limbs!!\nThe Knight replies, I\'ll knaw your ankles off!!\nYou advance on your quest leaving the Knight behind.')
        time.sleep(3)
        print('As you follow the path you evetually emerge from the forest.')
        time.sleep(3)
        print('You see before you a partially crumbled old stone bridge that looks as if unused for a long time.')
        time.sleep(5)
        bridge_crossing()

    time.sleep(3)
    print(f'Congratulations, Sir {player_name} you have complete your quest! ')

if __name__ == "__main__": 
    main()
```
